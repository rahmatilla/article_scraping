```python
from bs4 import BeautifulSoup
import pandas as pd
import datetime
import requests
import string
import re
```

```python
source = requests.get('https://kun.uz/uz/news/2022/11/30/parda-ortidagi-xususiylashtirish-yirik-aktivlar-qanday-qilib-ofshor-kompaniyalarga-otib-ketmoqda').text
soup = BeautifulSoup(source, 'lxml')
text=''
for p in soup.find_all("p"):
    text+=p.get_text()+" "
print(text)
```

```python
text = text.replace('”', "")
text = text.replace('“', "")
text = re.sub(r'[0-9]', '', text)
[word.strip(string.punctuation) for word in text.split()]
```
